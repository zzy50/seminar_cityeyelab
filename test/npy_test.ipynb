```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 10000000)
y = np.sqrt(25 - x**2)

plt.plot(x, y, color='blue')  # Set color to blue
plt.plot(x, -y, color='blue')  # Set color to blue
plt.xlabel('x')
plt.ylabel('y')
plt.title('Circle Plot')
plt.grid(True)  # Set aspect ratio to 1:1
plt.gca().set_aspect('equal')
plt.show()
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

r = 5

theta = np.linspace(0, 2 * np.pi, 50)
phi = np.linspace(0, np.pi, 50)

Theta, Phi = np.meshgrid(theta, phi)

X = r * np.sin(Phi) * np.cos(Theta)
Y = r * np.sin(Phi) * np.sin(Theta)
Z = r * np.cos(Phi)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X, Y, Z)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Scatter Plot of a Sphere')
plt.gca().set_aspect('equal')

plt.show()
```

```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

n_points = 1000

min_radius = 4
max_radius = 5

theta = np.random.uniform(0, 2 * np.pi, n_points)
# phi = np.random.uniform(0, np.pi, n_points)
phi = np.arccos(np.random.uniform(-1, 1, n_points))
r = np.random.uniform(min_radius, max_radius, n_points)

x = r * np.sin(phi) * np.cos(theta)
y = r * np.sin(phi) * np.sin(theta)
z = r * np.cos(phi)

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')

# ax.scatter(x, y, z)
ax.scatter(x - x.mean(), y - y.mean(), z - z.mean())
ax.scatter(0, 0, 0, color='red',  s=50)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Random Distribution on a Sphere')

plt.gca().set_aspect('equal')
plt.show()
```
